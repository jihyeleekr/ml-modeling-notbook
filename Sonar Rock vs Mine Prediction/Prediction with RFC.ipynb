```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('dataset.csv', header=None)
df.head()
```

```python
# Check the number of rows and columns
print(df.shape)
```

```python
# Statistical measures of the data
print(df.describe())
```

```python
# Check number of Mines and Rocks
'''
M -> Mine
R -> Rock
'''
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

```python
# Separate data and labels
X = df.drop(columns=60, axis=1)
Y= df[60]

print(X, Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 1)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Train the Random Forest Classification model with training data
model = RandomForestClassifier(
    n_estimators = 100,
    max_depth = None,
    random_state=42
)
model.fit(X_train, Y_train)
```

```python
# Check the performace of the model with accuracy score
train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(train_prediction, Y_train)
print("Accuracy Score for training data:", train_accuracy)
```

```python
# Check the performance of the model on test data
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, Y_test)
print("Accuracy Score for test data", test_accuracy)
```

## Visualization

```python
# Make prediction
Y_pred = model.predict(X_test)
cm = confusion_matrix(Y_test,Y_pred)

# Plot
plt.figure(figsize=(6,5))
sns.heatmap(cm,
           annot = True,
           fmt='d',
           cmap='Blues',
           xticklabels=['Rock', 'Mine'],
           yticklabels=['Rock', 'Mine'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regression Confusion Matrix')
plt.show()

print(classification_report(Y_test,Y_pred))
```
