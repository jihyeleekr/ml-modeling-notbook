```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
# Import the California Housing Data
housing = fetch_california_housing()
print(housing.data.shape, housing.target.shape)
print(housing)
```

```python
# Loading the dataset to a Pandas DataFrame
df = pd.DataFrame(housing.data, columns = housing.feature_names)
df.head()
```

```python
# Add the target (MedHouseVal in $100,000) column to the DF
df['MedHouseVal'] = housing.target
df.head()
```

```python
# Checking the number of rows and columns in the DF
df.shape
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Statistical measures of the dataset
df.describe()
```

```python
# Understanding the correleation between various features in the dataset
correlation = df.corr()
plt.figure(figsize =(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={'size':8}, cmap="Blues")
```

```python
# Split the data and Target
X = df.drop(['MedHouseVal'], axis = 1)
Y = df['MedHouseVal']
print(X)
print(Y)
```

```python
# Splitting the data into Train and Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
# Model Training with XGBoost Regessor
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
# Evaluation
training_prediction = model.predict(X_train)
# R squared error
score_1 = metrics.r2_score(Y_train, training_prediction)
# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_prediction)

print("R squared erro:", score_1)
print("Mean absolute error:", score_2)
```

```python
# Prediction on Test data
# Evaluation
test_prediction = model.predict(X_test)
# R squared error
score_11 = metrics.r2_score(Y_test, test_prediction)
# Mean Absolute Error
score_21 = metrics.mean_absolute_error(Y_test, test_prediction)

print("R squared erro:", score_11)
print("Mean absolute error:", score_21)
```
